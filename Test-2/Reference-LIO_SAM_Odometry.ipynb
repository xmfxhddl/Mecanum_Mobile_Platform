```python
import pickle
import matplotlib.pyplot as plt
```

```python
# Load the Reference Data
with open('./[실험값]Reference_Odometry_XY/edit_2.pickle', 'rb') as f:
    reference_data = pickle.load(f)

with open('./[실험값]LiDAR_Inertial_Odomerty_XY\edit_2.pickle', 'rb') as f:
    sensor_data = pickle.load(f)
```

```python
reference_x, reference_y = reference_data[:, 0], reference_data[:, 1]
sensor_x, sensor_y = sensor_data[:, 0], sensor_data[:, 1]
```

```python
# Create the plot
plt.figure(figsize=(9, 9), dpi=500)

# Plot the Reference Data
plt.plot(reference_x, reference_y, 'k-', label='Reference Data')

# Plot the Sensor Data as a line
plt.plot(sensor_x, sensor_y, 'r-', label='LIO SAM')

# Plot the start point
plt.plot(reference_x[0], reference_y[0], 'y*', label="Refernce data's Start Point")
plt.plot(sensor_x[0], sensor_y[0], 'b*', label="LIO SAM's Start Point")

# Plot the end point
plt.plot(reference_x[-1], reference_y[-1], 'ys', label="Refernce data's End Point")
plt.plot(sensor_x[-1], sensor_y[-1], 'bs', label="LIO SAM's End Point")

# Add labels and title
plt.axis([-7, 7, -2, 7])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Reference-LIO_SAM Trajectory')
plt.legend()

# Save the plot
save_path = './[결과]Plot_image/Reference-LIO_SAM Trajectory.png'
plt.savefig(save_path)

# Show the plot
plt.show()
```
